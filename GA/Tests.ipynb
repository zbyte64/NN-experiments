```python
from population import initialize_world
from operations import ReprFromMatrix, OperationsFromMatrix, OPERATIONS
import tensorflow as tf

assert 'conv2d' in OPERATIONS

import random

random.seed()

world_shape = 10,10
layer = tf.placeholder('float32', (None, 29, 29, 1))
world = initialize_world(world_shape, genome_shape=(6,5))
conv2d_examples = []
for i in range(world_shape[0]):
    for j in range(world_shape[1]):
        r = ReprFromMatrix(world[i,j], layer, include_all=True)
        print(r)
        for ct, ca, ck in r:
            if ct == 'conv2d':
                conv2d_examples.append(world[i,j])
                print(i,j,r)
                f = OperationsFromMatrix(layer, world[i,j])
                print(f)
                break
```
